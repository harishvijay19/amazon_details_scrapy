```python
import requests
from scrapy.http import TextResponse
```

```python
url="https://www.amazon.in/s/ref=sr_in_-2_p_89_1?fst=as%3Aoff&rh=n%3A14647040031%2Cp_89%3AApple&bbn=14647040031&ie=UTF8&qid=1535354816&rnid=3837712031"

user_agent = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/58: .0.3029.110 Chrome/58.0.3029.110 Safari/537.36'}
r=requests.get(url,headers=user_agent)
response=TextResponse(r.url,body=r.text,encoding='utf-8')
print(url)
```

```python
d=response.xpath('//*[@id="a-page"]//text()').extract()
print(d)
```

```python
c=response.xpath('//*[@id="result_0"]//text()').extract()
print("there are {}".format(len(c)))
print(c)
```

```python
import re
regex = (r'^[-+]?[0-9]+$')
print(regex)
matches = re.findall(regex,"you save")
for i in matches:
    resultI='\"'+"result_"+str(i)+'\"'
    c=response.xpath('//*[@id="+resultI+"]//text()').extract
    print(resultI)
print(c)
```

```python
for i in '\d[0-9]':
    resultI='\"'+"result_"+str(i)+'\"'
    print(resultI)

    c=response.xpath('//*[@id='+resultI+']//text()').extract()
    print("there are {}".format(len(c)))
    print(c)
```

```python
product={}
for i in range(0,9):
    resultI='\"'+"result_"+str(i)+'\"'
    print(resultI)
    #product['product_list1']=response.xpath('//*[@id='+resultI+']/div//text()').extract()
    product['product_name']=response.xpath('//*[@id='+resultI+']/div/div[3]/div[1]/a/h2//text()').extract()
    product['orginial_price']=response.xpath('//*[@id='+resultI+']/div/div[5]/div[1]/span[2]/text()').extract()
    product['reduced_price']=response.xpath('//*[@id='+resultI+']/div/div[5]/div[1]/a/span[2]/text()').extract()
    product['saved']=response.xpath('//*[@id='+resultI+']/div/div[5]/div[2]/span[2]/text()').extract()
    product['left_stock']=response.xpath('//*[@id='+resultI+']/div/div[5]/div[4]/div/span').extract()    
    product['star']=response.xpath('//*[@id='+resultI+']/div/div[7]/span/span/a/i[1]/span').extract()
    product['photo']=response.xpath('//*[@id='+resultI+']/div/div[2]/div/div/a/img').extract()
    print(product)
```

```python
import csv
csv_columns = ['product_name','orginal_price','reduced_price','saved','left_stock','star','photo']
dict_data = [{'product'}]
csv_file = "Names.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(dict_data.items())
except IOError:
    print("I/O error") 
```
